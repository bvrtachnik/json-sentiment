````python
# Load json file and extract lyrics
with open("birdhouse_lyrics.json", "r") as f:
    lyrics_data = json.load(f)

lyrics_text = "\n".join(lyrics_data["lyrics"])

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("spacytextblob")


# Run sentiment analysis on the lyrics
doc = nlp(lyrics_text)

# Print the lyrics
print("Lyrics:\n")
print(lyrics_text)

# Print the sentiment analysis results
print("\nSentiment Analysis Results:", doc._.blob.sentiment)

# The polarity is slightly positive, so the lyrics have a mostly neutral but positive tone.
````

### Question 3. 
- Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

````python
import lyricsgenius
import json

# Initialize Genius API client (use your own API key)
genius = lyricsgenius.Genius("qWhVrp9VgVJ1hLFW1iD7JjOGPKi7hX-bFL6kyG0iLOnDoG1S4MTGEqfXm6KpC7nm")

# Fetch lyrics for a song and save to JSON file
def save_lyrics_to_file(artist, song_title, filename):
    song = genius.search_song(song_title, artist)
    if song:
        # Split the lyrics by line and wrap in a dictionary
        result = {"lyrics": song.lyrics.split("\n")}
    
        # Save the result to JSON file
        with open(filename, "w") as f:
            json.dump(result, f, indent=4)
        print(f"Saved lyrics to {filename}")
    else:
        print(f"Lyrics not found for {song_title} by {artist}")

# Run the function for different songs
save_lyrics_to_file("Led Zeppelin", "Stairway to Heaven", "stairway.json")
save_lyrics_to_file("Jimi Hendrix", "All Along the Watchtower", "watchtower.json")
save_lyrics_to_file("Aerosmith", "Dream On", "dream_on.json")
save_lyrics_to_file("Boston", "More Than a Feeling", "feeling.json")
````
